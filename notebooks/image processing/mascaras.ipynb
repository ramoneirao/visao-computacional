```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

imagem = cv2.imread('../images/veropeso.jpg')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB) # O OpenCV armazena a imagem no formato BGR em vez de RGB
```

```python
# Masking 

# O mascaramento é uma técnica de segurança que envolve ocultar ou proteger informações sensíveis, como números de cartão de crédito, senhas ou dados pessoais, para evitar o acesso não autorizado. 
# O mascaramento pode ser feito de várias maneiras, como substituição de caracteres, criptografia ou ofuscação.
 
mask = np.zeros(imagem_rgb.shape[:2], dtype="uint8") # Cria uma máscara preta do mesmo tamanho da imagem
(cX, cY) = (imagem_rgb.shape[1] // 2, imagem_rgb.shape[0] // 2) # Define o centro da imagem
cv2.rectangle(mask, (cX - 550, cY - 280), (cX + 530, cY + 280), 255, -1) 

# Aplica a máscara na imagem
imagem_masked = cv2.bitwise_and(imagem_rgb, imagem_rgb, mask=mask) # Aplica a máscara na imagem

# Mácara circular com o raio de 100 pixels
mask_circular = np.zeros(imagem_rgb.shape[:2], dtype="uint8") # Cria uma máscara circular
cv2.circle(mask_circular, (cX, cY), 450, 255, -1) 
segunda_mascara = cv2.bitwise_and(imagem_rgb, imagem_rgb, mask=mask_circular) # Aplica a máscara circular na imagem
```

```python
plt.imshow(imagem_masked)
plt.show()
```

```python
plt.imshow(segunda_mascara)
plt.show()
```

```python
# Juntando as duas máscaras
imagem_final = cv2.add(imagem_masked, segunda_mascara) # Adiciona as duas máscaras
plt.imshow(imagem_final)
plt.show()
```
