```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
imagem = cv2.imread('../images/veropeso.jpg')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem_rgb)
plt.show()
```

```python
# Operações aritméticas com imagens

matriz = np.ones(imagem_rgb.shape, dtype="uint8") * 100 # cria uma matriz de mesma dimensão da imagem, com cada pixel igual a 100
soma = cv2.add(imagem_rgb, matriz) # soma a imagem com a matriz
plt.imshow(soma)
plt.show()
```

```python
matriz = np.ones(imagem_rgb.shape, dtype="uint8") * 50
subtracao = cv2.subtract(imagem_rgb, matriz) # subtrai a matriz da imagem
plt.imshow(subtracao)
plt.show()
```

```python
# Operações lógicas com imagens

# Desenhando um retângulo
fundo = np.zeros((300, 300), dtype="uint8") # cria uma imagem preta
cv2.rectangle(fundo, (25, 25), (275, 275), 255, -1) # desenha um retângulo branco

# Desenhando um círculo
fundo2 = np.zeros((300, 300), dtype="uint8") # cria uma imagem preta
cv2.circle(fundo2, (150, 150), 150, 255, -1) # desenha um círculo branco

# Operações lógicas
bitwise_and = cv2.bitwise_and(fundo, fundo2) # operação AND
bitwise_or = cv2.bitwise_or(fundo, fundo2) # operação OR
bitwise_xor = cv2.bitwise_xor(fundo, fundo2) # operação XOR
bitwise_not = cv2.bitwise_not(fundo2) # operação NOT
```

```python
plt.imshow(fundo)
plt.show()
plt.imshow(fundo2)
plt.show()        
```

```python
plt.subplot(221), plt.imshow(bitwise_and)
plt.title('AND'), plt.xticks([])
plt.subplot(222), plt.imshow(bitwise_or)
plt.title('OR'), plt.xticks([])
plt.subplot(223), plt.imshow(bitwise_xor)
plt.title('XOR'), plt.xticks([])
plt.subplot(224), plt.imshow(bitwise_not)
plt.title('NOT'), plt.xticks([])
plt.show()
```

```python
# Operações morfológicas (erosão, dilatação, opening, closing e gradiente)
# Erosão
# É a técnica de remover pixels perto dos limites de um objeto, com o objetivo de ressaltar esse objeto.
# A dilatação é o oposto da erosão, ou seja, ela adiciona pixels aos limites de um objeto.
#
# Na erosão, o kernel de convolução substitui o valor central pelo menor valor encontrado no kernel.
# Objetivo: preservar apenas regiões de alta intensidade, reduzindo imediatamente regiões que não são completas em intensidade.
#
# Na dilatação, o kernel de convolução substitui o valor central pelo maior valor encontrado no kernel.
# Objetivo: aumentar regiões que não são completas em intensidade.

# Aplicando erosão
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY) 
eroded = cv2.erode(imagem_cinza.copy(), (5, 5), iterations=3) # (imagem, kernel, iterações)
cv2.imshow('Imagem Cinza', imagem_cinza)
cv2.imshow('Erosão', eroded)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Mostrando a imagem original e a imagem erodida
plt.figure(figsize=(15, 10))
plt.subplot(121), plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(eroded, cmap='gray')
plt.title('Erosão'), plt.xticks([]), plt.yticks([])
plt.show()
```
