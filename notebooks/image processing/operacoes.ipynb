```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
imagem = cv2.imread('../images/veropeso.jpg')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.imshow(imagem_rgb)
plt.show()
```

```python
# Operaes aritmticas com imagens

matriz = np.ones(imagem_rgb.shape, dtype="uint8") * 100 # cria uma matriz de mesma dimenso da imagem, com cada pixel igual a 100
soma = cv2.add(imagem_rgb, matriz) # soma a imagem com a matriz
plt.imshow(soma)
plt.show()
```

```python
matriz = np.ones(imagem_rgb.shape, dtype="uint8") * 50
subtracao = cv2.subtract(imagem_rgb, matriz) # subtrai a matriz da imagem
plt.imshow(subtracao)
plt.show()
```

```python
# Operaes lgicas com imagens

# Desenhando um retngulo
fundo = np.zeros((300, 300), dtype="uint8") # cria uma imagem preta
cv2.rectangle(fundo, (25, 25), (275, 275), 255, -1) # desenha um retngulo branco

# Desenhando um crculo
fundo2 = np.zeros((300, 300), dtype="uint8") # cria uma imagem preta
cv2.circle(fundo2, (150, 150), 150, 255, -1) # desenha um crculo branco

# Operaes lgicas
bitwise_and = cv2.bitwise_and(fundo, fundo2) # operao AND
bitwise_or = cv2.bitwise_or(fundo, fundo2) # operao OR
bitwise_xor = cv2.bitwise_xor(fundo, fundo2) # operao XOR
bitwise_not = cv2.bitwise_not(fundo2) # operao NOT
```

```python
plt.imshow(fundo)
plt.show()
plt.imshow(fundo2)
plt.show()        
```

```python
plt.subplot(221), plt.imshow(bitwise_and)
plt.title('AND'), plt.xticks([])
plt.subplot(222), plt.imshow(bitwise_or)
plt.title('OR'), plt.xticks([])
plt.subplot(223), plt.imshow(bitwise_xor)
plt.title('XOR'), plt.xticks([])
plt.subplot(224), plt.imshow(bitwise_not)
plt.title('NOT'), plt.xticks([])
plt.show()
```

```python
# Operaes morfolgicas (eroso, dilatao, opening, closing e gradiente)
# Eroso
#  a tcnica de remover pixels perto dos limites de um objeto, com o objetivo de ressaltar esse objeto.
# A dilatao  o oposto da eroso, ou seja, ela adiciona pixels aos limites de um objeto.
#
# Na eroso, o kernel de convoluo substitui o valor central pelo menor valor encontrado no kernel.
# Objetivo: preservar apenas regies de alta intensidade, reduzindo imediatamente regies que no so completas em intensidade.
#
# Na dilatao, o kernel de convoluo substitui o valor central pelo maior valor encontrado no kernel.
# Objetivo: aumentar regies que no so completas em intensidade.

# Aplicando eroso
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY) 
eroded = cv2.erode(imagem_cinza.copy(), (5, 5), iterations=3) # (imagem, kernel, iteraes)
cv2.imshow('Imagem Cinza', imagem_cinza)
cv2.imshow('Eroso', eroded)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Mostrando a imagem original e a imagem erodida
plt.figure(figsize=(15, 10))
plt.subplot(121), plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(eroded, cmap='gray')
plt.title('Eroso'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Aplicando dilatao
dilated = cv2.dilate(imagem_cinza.copy(), None, iterations=3) # (imagem, kernel, iteraes) kernel padro 3x3
cv2.imshow('Imagem Cinza', imagem_cinza)
cv2.imshow('Dilatao', dilated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Mostrando a imagem original e a imagem dilatada
plt.figure(figsize=(15, 10))
plt.subplot(121), plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(dilated, cmap='gray')
plt.title('Dilatao'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Aplicando opening
# O opening  uma operao que remove pequenos objetos da imagem, preservando a forma e o tamanho dos objetos maiores.
#  uma combinao de eroso seguida de dilatao.
imagem_ruido = cv2.imread('../images/ruido.jpg')
opened = cv2.morphologyEx(imagem_ruido, cv2.MORPH_OPEN, None) 
cv2.imshow('Imagem com Rudo', imagem_ruido)
cv2.imshow('Opening', opened)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Mostrando a imagem com ruido e a imagem com opening
plt.figure(figsize=(15, 10))
plt.subplot(121), plt.imshow(imagem_ruido, cmap='gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(opened, cmap='gray')
plt.title('Opening'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Aplicando closing
# O closing  uma operao que preenche pequenos buracos na imagem, preservando a forma e o tamanho dos objetos maiores.
#  uma combinao de dilatao seguida de eroso.
closed = cv2.morphologyEx(imagem_ruido, cv2.MORPH_CLOSE, None)  
cv2.imshow('Imagem com Rudo', imagem_ruido)
cv2.imshow('Closing', closed)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Mostrando a imagem com ruido e a imagem com opening
plt.figure(figsize=(15, 10))
plt.subplot(121), plt.imshow(imagem_ruido, cmap='gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(closed, cmap='gray')
plt.title('Closing'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Morphological Gradient
# O gradiente morfolgico  a diferena entre a dilatao e a eroso da imagem.
# Ele destaca as bordas dos objetos na imagem.
gradient = cv2.morphologyEx(imagem_cinza, cv2.MORPH_GRADIENT, None)
cv2.imshow('Imagem com Rudo', imagem_cinza)
cv2.imshow('Gradient', gradient)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Mostrando a imagem com ruido e a imagem com opening
plt.figure(figsize=(15, 10))
plt.subplot(121), plt.imshow(imagem_cinza, cmap='gray')
plt.title('Imagem Original'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(gradient, cmap='gray')
plt.title('Gradient'), plt.xticks([]), plt.yticks([])
plt.show()
```
