```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Desenhando imagens com o opencv
# Inicializando nossa tela como um espaço de pixels 500x500 com 3 canais (RGB)
espaco = np.zeros((500, 500, 3), dtype = 'uint8')

# Desenhando uma linha verde no canto superior esquerdo da nossa tela para o canto inferior direito
verde = (0, 255, 0)
cv2.line(espaco, (0, 0), (500, 500), verde)
cv2.imshow('Figura', espaco)
cv2.waitKey(0)

# Desenhando uma linha vermelha de 3 pixels de espessura do canto inferior esquerdo para a parte superior direita 
vermelho = (0, 0, 255)
cv2.line(espaco, (500, 0), (0, 500), vermelho, 3)
cv2.imshow('Figura', espaco)
cv2.waitKey(0)

# Desenhando uma quadrado verde de 40x40 pixels, começando em 20x20 e termina em 60x60 
cv2.rectangle(espaco, (20, 20), (60, 60), verde)
cv2.imshow('Figura', espaco)
cv2.waitKey(0)

# Desenhando outro retangulo, vermelho e com 5 pixels de espessura 
cv2.rectangle(espaco, (10, 150), (100, 300), vermelho, 5)
cv2.imshow('Figura', espaco)
cv2.waitKey(0)

# Desenhando retangulo, azul e preenchido 
azul = (255, 0, 0)
cv2.rectangle(espaco, (400, 80), (480, 160), azul, -1)
cv2.imshow('Figura', espaco)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(espaco)
plt.show()
```

```python
# Desenhando um círculo branco no centro da tela com vários circulos ao redor
espaco1 = np.zeros((500, 500, 3), dtype = 'uint8')
(centro_x, centro_y) = (espaco1.shape[1]//2, espaco1.shape[0]//2) # Localização x, y do centro
branco = (255, 255, 255)

for raio in range(0, 250, 25):
    cv2.circle(espaco1, (centro_x, centro_y), raio, branco)

cv2.imshow('Figura', espaco1)
cv2.waitKey(0)
```

```python
plt.imshow(espaco1)
plt.show()
```
