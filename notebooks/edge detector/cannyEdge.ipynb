```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
imagem = cv2.imread('../images/veropeso.jpg')
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
```

```python
# Canny Edge Detection
imagem_com_filtro = cv2.GaussianBlur(imagem_cinza, (5, 5), 0) # aplicando um filtro gaussiano para ruído

# Mostrando a imagem desfocada
# cv2.imshow('Imagem desfocada', imagem_com_filtro)
# cv2.waitKey(0)

# Calculando os filtros Canny com diferentes limiares
largo = cv2.Canny(imagem_com_filtro, 50, 220)
medio = cv2.Canny(imagem_com_filtro, 70, 140)
estreito = cv2.Canny(imagem_com_filtro, 210, 220)

# Mostrando os mapas de bordas (map edges) com os diferentes limiares
cv2.imshow('Canny - Largo', largo)
cv2.waitKey(0)
cv2.imshow('Canny - Médio', medio)
cv2.waitKey(0)
cv2.imshow('Canny - Estreito', estreito)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Mostrando os mapas de bordas (map edges) com os diferentes limiares
plt.figure(figsize=(20, 10))
plt.subplot(131), plt.imshow(largo, cmap='gray')
plt.title('Largo'), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(medio, cmap='gray')
plt.title('Medio'), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(estreito, cmap='gray')
plt.title('Estreito'), plt.xticks([]), plt.yticks([])
plt.show()
```
